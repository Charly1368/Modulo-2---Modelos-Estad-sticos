```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el experimento aleatorio
# Espacio muestral: Lanzamiento de un dado justo
R_X = np.array([1, 2, 3, 4, 5, 6])
P_X = np.array([1/6] * 6)  # Probabilidad uniforme

# Verificar que la suma de probabilidades es 1
suma_prob = np.sum(P_X)
print(f"Suma de probabilidades: {suma_prob}")  # Debería ser 1

# Graficar la función de densidad
plt.bar(R_X, P_X, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel("Valores de X")
plt.ylabel("Probabilidad P(X)")
plt.title("Función de Masa de Probabilidad (FMP)")
plt.xticks(R_X)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
